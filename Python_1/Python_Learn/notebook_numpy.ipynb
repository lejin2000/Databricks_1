```python
import numpy as np

arr = np.array([1, 2, 3, 4])
display(arr)  # Output: [1 2 3 4]
```

```python
#Multi-dimensional array:
matrix = np.array([[1, 2], [3, 4]])
print(matrix)
```

```python
#Arrays of zeros and ones:
zeros = np.zeros((4, 3))
ones = np.ones((1, 4))
print(zeros)
print(ones)
```

```python
#Array with a range of values:
range_arr = np.arange(5)  # Creates array [0 1 2 3 4]
print(range_arr)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)  # [5 7 9]
print(a * b)  # [4 10 18]

print(np.sum(a))       # Sum: 6
print(np.mean(b))      # Mean: 5.0
print(np.sqrt(a))      # Square root: [1. 1.414 1.732]
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([10, 20])
print(arr1 + arr2)
```

```python
a = np.array([1, 2])
b = np.array([3, 4])
print(np.vstack((a, b)))  # Vertical stack
# [[1 2]
#  [3 4]]

print(np.hstack((a, b)))  # Horizontal stack
# [1 2 3 4]
```
