```python
import pandas as pd
#create df with 2 column
df = pd.DataFrame(columns=["value","value2"])
print(df)
```

```python
#add another column
df["value3"] = []
print(df)
```

```python
# Insert a new row at the end with index = len(df)
new_row = [1, 2, 3]
df.loc[len(df)] = new_row
print(df)
```

```python
# Delete row with index 1
df = df.drop(1)

print(df)
```

```python
#set value as index
df = df.set_index('value')
print(df)
```

```python
df.loc[2] = ['Charlie', 35]
print(df)
```

```python
df["value5"] = ["",""]
print(df)
```

```python
# Drop column 'B'
column_to_drop = 'value5'  # Example column, may or may not exist

if column_to_drop in df.columns:
    df = df.drop(column_to_drop, axis=1)
else:
    print(f"Column '{column_to_drop}' not found, skipping drop.")
```

```python
df.loc[3] = ['Charlie3', 35,"XX"]
print(df)
```

```python
# Insert blank row at the end (index = length of DataFrame)
df.loc[len(df)] = [None] * len(df.columns)
print(df)
```

```python
# Insert blank row at the end (index = length of DataFrame)
df.loc["A"] = [None] * len(df.columns)
print(df)
```
