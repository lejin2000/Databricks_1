```python
# Create a dictionary
student = {"name": "Alice", "age": 22, "course": "Computer Science"}
```

```python
# Add or update a key-value pair
student["age"] = 23

# Use setdefault to get or add a key
print(student.setdefault("grade", "A"))  # Adds "grade": "A" if not present
```

```python
print(student["name"])  # Output: Alice
```

```python
for key, value in student.items():
    print(key, value)
```

```python
# Create a set
fruits = {"apple", "banana", "cherry"}

# Add an item to the set
fruits.add("orange")

# Attempt to add a duplicate (ignored)
fruits.add("apple")
```

```python
# Print set
print(fruits)  # Output: {'banana', 'orange', 'apple', 'cherry'}
```

```python
# Set operations
a = {1, 2, 3}
b = {3, 4, 5}
print(a.union(b))        # {1, 2, 3, 4, 5}
print(a.intersection(b)) # {3}
print(a.difference(b))   # {1, 2}
```
