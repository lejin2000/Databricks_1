```python
df_csv = spark.read.csv(
    "/Volumes/databricks_simulated_retail_customer_data/v01/source_files/sales.csv",
    header=True,
    inferSchema=True
)

display(df_csv)
```

```python
df_csv.createOrReplaceTempView("csv")
spark.sql("SELECT count(*) FROM csv").show()
```

```python
df_name = spark.sql("SELECT distinct customer_name from csv order by customer_name")
display(df_name)
```

```python
from pyspark.sql.functions import desc

df_productname = spark.sql("SELECT distinct product_name,count(1) as count from csv  group by product_name")
#df_productname = df_productname.orderBy("count", ascending=False) 
df_productname = df_productname.orderBy(desc("count"))
display(df_productname)
```

```python
val = df_productname.first()["product_name"]
#display(val)

df_allrows = spark.sql("SELECT * from csv where product_name = '" + val + "'")
display(df_allrows)
```
