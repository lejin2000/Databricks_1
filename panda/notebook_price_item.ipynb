```python
# If not already installed, do: pip install pandas fastparquet
import pandas as pd

URL_DATA = 'https://storage.data.gov.my/pricecatcher/lookup_item.parquet'

df_lookup_item = pd.read_parquet(URL_DATA)
print(df_lookup_item.count())
spark_df_lookup_item = spark.createDataFrame(df_lookup_item)   
spark_df_lookup_item.createOrReplaceTempView("vw_lookup_item")
#if 'date' in df.columns: df['date'] = pd.to_datetime(df['date'])

#display(df)
```

```python
%sql
SELECT *
FROM vw_price_catcher pc
INNER JOIN vw_lookup_premise lp ON pc.premise_code = lp.premise_code
INNER JOIN vw_lookup_item li ON pc.item_code = li.item_code
WHERE state  = 'W.P. Kuala Lumpur'
```

```python
result = spark.sql("""
SELECT pc.date,
pc.premise_code,
pc.item_code,
pc.price,
--lp.premise_code,
lp.premise,
lp.address,
lp.premise_type,
lp.state,
lp.district,
--li.item_code,
li.item,
li.unit,
li.item_group,
li.item_category
FROM vw_price_catcher pc
INNER JOIN vw_lookup_premise lp ON pc.premise_code = lp.premise_code
INNER JOIN vw_lookup_item li ON pc.item_code = li.item_code
""")

result.write.mode("overwrite").saveAsTable("dbacademy.default.price_catcher")
```
